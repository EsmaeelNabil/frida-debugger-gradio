```python
import frida
import gradio as gr


def apps():
    device = frida.get_usb_device()
    apps = device.enumerate_applications(scope="full")
    appslist = []
    for app in apps:
        params = dict(app.parameters)
        app = f' identifier="{app.identifier}", name="{app.name}", pid={app.pid}'
        appslist.append(app)
    new_string = ", ".join(appslist)
    return new_string

demo = gr.Interface(fn=apps,inputs="text",outputs="text")
demo.launch()
```
